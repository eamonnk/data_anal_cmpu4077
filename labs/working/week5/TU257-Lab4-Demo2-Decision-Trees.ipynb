```python
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
```

```python
# load the iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris
```

```python
X = iris.data
y = iris.target
```

```python
# Fit the classifier with default hyper-parameters
# defining random_state gives us reproducability
clf = DecisionTreeClassifier(random_state=1234)
model = clf.fit(X, y)
```

```python
# these are the different rules that were generated
# i.e. fi 3rd variable figure_2 is <= 2.45 then class will be zero
text_representation = tree.export_text(clf)
print(text_representation)
```

```python
# convert the tree output above into a graphical output of the decisions tree
# plot out the nodel alogng with names --- this is mofdel for those set of rules we saw above
# the terminations nodes where theres no branches are where predictions happen
target_name=['0','1']
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

```python
from sklearn import preprocessing

# encode categorical variables using label Encoder

# select all categorical variables
df_categorical = df.select_dtypes(include=['object'])
df_categorical.head()
```

```python
# apply label encoder to df_categorical
le = preprocessing.LabelEncoder()
df_categorical = df_categorical.apply(le.fit_transform)
df_categorical.head()
```

```python
# Next, Concatenate df_categorical dataframe with original df (dataframe)

# first, Drop earlier duplicate columns which had categorical values
df = df.drop(df_categorical.columns,axis=1)
df = pd.concat([df,df_categorical],axis=1)
df.head()
```

```python
corr_matrix=df.corr()
corr_matrix
```

```python
import seaborn as sn
#import matplotlib.pyplot as plt

fig = plt.subplots(figsize=(17,14))
sn.heatmap(corr_matrix, annot=True)
```

```python
df['Target'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

# Putting independent variables/features to X
# remove target variable
X = df.drop('Target',axis=1)

# Putting response/dependent variable/feature to y
# Target variable
y = df['Target']
```

```python
# Splitting the data into train and test
#30% test 70% train
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=99)
```

```python
# Importing decision tree classifier from sklearn library
from sklearn.tree import DecisionTreeClassifier

# Fitting the decision tree with default hyperparameters, apart from
# max_depth which is 5 so that we can plot and read the tree.
# coudl set randomstate if we want
# want to set depth  -howe mahy layers it will go to...
dt_default = DecisionTreeClassifier(max_depth=5)
dt_default.fit(X_train,y_train)
```

```python
text_representation = tree.export_text(dt_default)
print(text_representation)
```

```python
target_names=['0','1']
feature_names=['Age', 'WorkClass', 'Fnlwgt', 'Education', 'Edu_Num', 'MaritalStatus', 'Occupation', 'Relationship', 'Race', 'Sex', 'CapitalGain', 'CapitalLoss', 'HrPerWk', 'Native']
fig = plt.figure(figsize=(20,18))
_ = tree.plot_tree(dt_default, filled=True, proportion=True, fontsize=10, feature_names=feature_names, class_names=target_names)
```

```python
# can modifyt to go to max depth of 2 layers to make it easier to see if we wnat
# may or may not be possible depending on problem
fig = plt.figure(figsize=(20,18))
_ = tree.plot_tree(dt_default, filled=True, proportion=True, fontsize=10, max_depth=2, feature_names=feature_names, class_names=target_names)
```

```python
# how good is it - whats thew accuracy.... Classification Report below...
# Importing classification report and confusion matrix from sklearn metrics
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

# making predictions
y_pred_default = dt_default.predict(X_test)

# Printing classifier report after prediction
print(classification_report(y_test,y_pred_default))
```

```python
#Exercise: See previous example for creating a Confusion matrix.  Create a confusion matrix for th results
# I justmodified the y_test and y_pred_default values

import seaborn as sns
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

# passing actual and predicted values
cm = confusion_matrix(y_test, y_pred_default)

# true write data values in each cell of the matrix
sns.heatmap(cm, annot=True)
plt.title('Confusion Matrix', fontsize = 15) 
plt.xlabel('Predicted', fontsize = 13) 
plt.ylabel('Acuals', fontsize = 13) 

plt.show()
```
