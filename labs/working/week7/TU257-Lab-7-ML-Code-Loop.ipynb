```python
from sklearn import preprocessing

# encode categorical variables using label Encoder

# select all categorical variables
df_categorical = df.select_dtypes(include=['object'])
df_categorical.head()
```

```python
# apply label encoder to df_categorical
le = preprocessing.LabelEncoder()
df_categorical = df_categorical.apply(le.fit_transform)
df_categorical.head()
```

```python
# Next, Concatenate df_categorical dataframe with original df (dataframe)

# first, Drop earlier duplicate columns which had categorical values
df = df.drop(df_categorical.columns,axis=1)
df = pd.concat([df,df_categorical],axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Putting independent variables/features to X
X = df.drop('Target',axis=1)

# Putting response/dependent variable/feature to y
y = df['Target']
```

```python
# Splitting the data into train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=99)
```

#### Now we can setup and Loop through the different algorithm, gathering data and results with each iteration of the loop
#### 

```python
import pandas as pd
from sklearn import model_selection
from sklearn.metrics import accuracy_score, log_loss
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

log_cols=["Classifier", "Accuracy"]
log = pd.DataFrame(columns=log_cols)

# ste up list of dataframs for each classifier what is accuracy rate that we ahve achieved
# use it later
# what are modelswe'll use - listed to below - can add other models too
classifiers = [
   ('KNN', KNeighborsClassifier(3)),
   ('NB', GaussianNB()),
   ('DT', DecisionTreeClassifier()),
   ('RF', RandomForestClassifier()),
   ('LR', LogisticRegression())]
```

Create Models & Generate Confusion Matrix & ROC Chart

Create a Loop to process each algorithm in the list, calculate the accuracy, confusion matrix and values for creating a ROC chart.

```python
from sklearn.metrics import roc_curve, auc

plt.clf()
plt.figure(figsize=(18,13))

# this is core part of hwat we're doing
# step through
# first fit or train the model  - pull out name and model - loop through clssifier list from earlier to do so
# take model and train it
for name, model in classifiers:
   model.fit(X_train, y_train)
   name = model.__class__.__name__

   # then do predictions
   # predictions on test data set and calculate accuracy score from those and print them
   train_predictions = model.predict(X_test)
   acc = accuracy_score(y_test, train_predictions)
   print (name, "Accuracy=", acc)

   # take accuracy and store ity in our dataframne
   # take name * 100, and append to existing dataframe
   log_entry = pd.DataFrame([[name, acc*100]], columns=log_cols)
   # log = log.append(log_entry)
   log = pd.concat([log, log_entry])

   # roc chart
   # calculate prediction probability for test data set,  take poredicted probs formtest data set
   Y_predict_prob = model.predict_proba(X_test)[:,1]

   # calculate 
   fpr, tpr, thresholds = roc_curve(y_test, Y_predict_prob)
   roc_auc = auc(fpr, tpr)

   #generate data for confusion matrix
   print(pd.crosstab(pd.Series(y_test), np.array(train_predictions), rownames=['Actual'], colnames=['Predicted'], margins=True))
   print('')

   # add that to a plot - false positive and true positive - repeat through for loop
   plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % (name, roc_auc))

#PLots out the ROC chart
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc=0, fontsize='small')
plt.show()
```

```python
#Load the Seaborn library
import seaborn as sns

sns.set_color_codes("muted")
sns.barplot(x='Accuracy', y='Classifier', data=log, color="b")

plt.xlabel('Accuracy %')
plt.title('Classifier Accuracy')
plt.show()
```
